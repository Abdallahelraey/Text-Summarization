### Data Standerization

```python
import nltk
import re
from pathlib import Path
from typing import List, Optional, Set
import string
import os
from TextSummarizer.logging import logger

def setup_nltk_environment(data_dir: Optional[str] = None) -> str:
    if data_dir:
        nltk_data_dir = Path(data_dir)
        nltk_data_dir.mkdir(parents=True, exist_ok=True)
        
        # Set NLTK_DATA environment variable
        os.environ['NLTK_DATA'] = str(nltk_data_dir)
        
        # Clear existing paths and add custom path
        nltk.data.path = [str(nltk_data_dir)]
        return str(nltk_data_dir)
    return nltk.data.find('.').parent

def download_nltk_models(data_dir: Optional[str] = None, 
                        models: Optional[List[str]] = None) -> Set[str]:
    default_models = [
        'punkt',           # Sentence tokenizer
        'stopwords',       # Common stopwords
        'averaged_perceptron_tagger',  # POS tagger
        'wordnet',         # Lexical database
        'words'            # Word lists
    ]
    
    # Setup environment if custom directory provided
    if data_dir:
        setup_nltk_environment(data_dir)
    
    models_to_download = models if models else default_models
    downloaded_models = set()
    
    for model in models_to_download:
        try:
            nltk.download(model, 
                         download_dir=data_dir if data_dir else None,
                         quiet=True)
            downloaded_models.add(model)
            logger.info(f"Successfully downloaded '{model}' to {data_dir if data_dir else 'default location'}")
        except Exception as e:
            logger.info(f"Failed to download '{model}': {e}")
    
    return downloaded_models
```

```python
def clean_text(text: str) -> str:
    # Convert to lowercase
    text = text.lower()
    
    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = ' '.join(text.split())
    
    return text

def remove_stopwords(text: str, custom_stopwords: Optional[Set[str]] = None) -> str:
    try:
        from nltk.corpus import stopwords
        stop_words = set(stopwords.words('english'))
        if custom_stopwords:
            stop_words.update(custom_stopwords)
    except LookupError:
        print("Downloading stopwords...")
        nltk.download('stopwords')
        from nltk.corpus import stopwords
        stop_words = set(stopwords.words('english'))
    
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

def lemmatize_text(text: str) -> str:
    try:
        from nltk.stem import WordNetLemmatizer
        lemmatizer = WordNetLemmatizer()
    except LookupError:
        print("Downloading required models...")
        nltk.download('wordnet')
        from nltk.stem import WordNetLemmatizer
        lemmatizer = WordNetLemmatizer()
    
    words = text.split()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)

def standardize_text(text: str, 
                    remove_stops: bool = True,
                    lemmatize: bool = True,
                    custom_stopwords: Optional[Set[str]] = None) -> str:
    # Clean the text first
    text = clean_text(text)
    
    # Remove stopwords if requested
    if remove_stops:
        text = remove_stopwords(text, custom_stopwords)
    
    # Lemmatize if requested
    if lemmatize:
        text = lemmatize_text(text)
    
    return text
```

```python
# Specify custom directory
custom_dir = r"D:\AI\NLP\HandsOn\Text Summarization\artifacts\models"

# Setup NLTK environment with custom directory
nltk_dir = setup_nltk_environment(custom_dir)
print(f"NLTK data directory: {nltk_dir}")

# Download required models to custom directory
downloaded = download_nltk_models(data_dir=custom_dir)
print(f"Downloaded models: {downloaded}")

# Example text
sample_text = "The quick brown fox's jump wasn't particularly high! However, it was jumping over lazy dogs."

# Test the functions
print("\nOriginal text:", sample_text)
print("\nCleaned text:", clean_text(sample_text))
print("\nText without stopwords:", remove_stopwords(clean_text(sample_text)))
print("\nLemmatized text:", lemmatize_text(clean_text(sample_text)))
print("\nFully standardized text:", standardize_text(sample_text))
```
